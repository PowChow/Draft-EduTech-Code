```python
#KBAI
kbai_nodes = pd.read_csv('./user_data/final/KBAI_Nodes_all.csv')
kbai_survey = pd.read_csv('./user_data/final/KBAI_Surveys.csv')

print(len(kbai_nodes), len(kbai_survey))
kbai_final = kbai_nodes.join(kbai_survey)
kbai_final.to_csv('./user_data/final/KBAI_combined_v1.csv')
```

```python
#EDUTECH
et_nodes = pd.read_csv('./user_data/final/Edu_Nodes_all.csv')
et_survey = pd.read_csv('./user_data/final/Edu_Surveys.csv')

print(len(et_nodes), len(et_survey))
et_final = et_nodes.join(et_survey)
et_final.to_csv('./user_data/final/EDUTECH_combined_v1.csv')
```

```python
#graphs 
g = sns.FacetGrid(et_final, row="gender", col="grade_bucket", margin_titles=True)
bins = np.linspace(0, 60, 13)
g.map(plt.hist, "days", color="steelblue", bins=bins, lw=0)
```
