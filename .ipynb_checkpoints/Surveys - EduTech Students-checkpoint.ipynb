```python
# Question from Quarter Course Survey
# Personal Performance [1,2]
# Class Review [3-13, 15-18]
s2 = pd.read_csv('./2015EdTech/Assessment-Quarter-Course_Survey-03302016.txt', sep='\t')
#s2.columns = ['quarter_' + c for c in s2.columns]
```

```python
s3 = pd.read_csv('./2015EdTech/Assessment-Mid-Course_Survey-03302016.txt', sep='\t')
#s3.columns = ['midcourse_' + c for c in s3.columns]

s3.head(2)
```

```python
s4 = pd.read_csv('./2015EdTech/Assessment-End_of_Course_Survey-03302016.txt', sep='\t')
#s4.columns = ['end_' + c for c in s4.columns]
```

```python
s4.head()
```

```python
tmp = s2.ix[:, 2:].merge(s3.ix[:, 2:], on='User Name', suffixes=('_quarter', '_mid'))
surveys  = tmp.merge(s4, on='User Name')
surveys 
```

```python
surveys.columns
```

```python
#compare student responses through different surveys
student_confidence = surveys.ix[:,['User Name', 'Question 1 text_quarter', 'Question 2 text_quarter',
              'Question 1 text_mid', 'Question 2 text_mid', 'Question 10 text']]

student_confidence.columns = ['User Name', 'Quarter Performance So Far', 'Quarter Confidence Suceed in Class',
                   'Mid Performance So Far', 'Mid Confidence Suceed in Class', 'End Class Performance' ]

student_confidence
```

```python
#Test 
text_str = "2 - Somewhat confident"
text_str = re.sub(r"\D", "", text_str)
text_str
```

```python
def get_digits(x):
    #print x
    val = re.sub(r"\D", "", str(x))
    #print val
    if val: 
        return int(val)
    else:
        return val
```

```python
for c in student_confidence.ix[:, 1:].columns:
    student_confidence.ix[:, c] = student_confidence.ix[:, c].apply(lambda x: get_digits(x))
```

```python
# Change in confidency by student
s_performance = student_confidence.ix[:, ['User Name', 'Quarter Performance So Far', 'Mid Performance So Far', 'End Class Performance']]
s_performance.to_csv('./2015EdTech/Survey_Student_Performance.csv')
```
